```python
import pandas
import synapseclient

syn = synapseclient.login(silent=True)
```

```python
e = syn.getEvaluation(4484862)
```

```python
regListObj = syn.get('syn4550116')
regList = pandas.read_csv(regListObj.path)
```

```python
sb = list(syn.getSubmissionBundles(e, status="VALIDATED"))
```

```python
teams = dict()

for (submission, status) in sb:
    try:
        teamId = submission.teamId
        teamMembers = [(teamId, x.member.userName) for x in syn.getTeamMembers(teamId)]
    except:
        teamId = submission.userId
        teamMembers = [(teamId, syn.getUserProfile(teamId).userName)]
    
    teams[submission.id] = teamMembers
```

```python
teams = pandas.DataFrame([y for x in teams.itervalues() for y in x])
teams.columns = ["teamId", "userName"]

teams = teams.merge(regList, how="left").drop_duplicates()
```
